```python
import json
import os
import pandas as pd
import numpy as np
```

```python
path = os.getcwd()
path += '/tweetsData'
FileNames = []
for filename in os.listdir(path):
    FileNames.append(filename)
```

```python
tweets = []
for file in FileNames:
    with open(file, 'r') as f:
        for line in f.readlines():
            tweets.append(json.loads(line))
```

```python
def populate_tweet_df(tweets):
    df = pd.DataFrame()
    df['text'] = list(map(lambda tweet: tweet['text'], tweets))
    return df
```

```python
data = populate_tweet_df(tweets)
```

```python
tweet_words = []
for line in data['text']:
    twetwords = line.split(" ")
    for word in twetwords:
            if '#' in word:
                tweet_words.append(word[1:])
            else:
                tweet_words.append(word)
```

```python
word_files = ['FullDict/positive.txt','FullDict/negative.txt']
words = {}
for file in word_files:
    with open(file, 'r') as f:
        dict_name = file.split('.')[0].split("/")
        word = []
        for line in f.readlines():
            line = line.strip()
            if '*' in line:
                word.append(line[:-1])
            else:
                word.append(line)
        words[dict_name[1]] = word
```

```python
Positive, Negative = [], []
for word in tweet_words:
    for typ in words:
        word = word.upper()
        if word in words[typ]:
            if typ == 'positive':
                Positive.append(word)
            else:
                Negative.append(word)
```

```python
Negative = np.unique(Negative)
```

```python
Positive = np.unique(Positive)
```

```python
with open('Dictionary/positive.txt', 'w') as ptr:
    for item in Positive:
        ptr.write("%s\n" %item)
with open('Dictionary/negative.txt', 'w') as ptr:
    for item in Negative:
        ptr.write("%s\n" %item)
```
